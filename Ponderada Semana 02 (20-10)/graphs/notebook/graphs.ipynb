```python
import matplotlib.pyplot as plt
import numpy as np

# Dados fornecidos
dados = """
4825 5.00 0.00
5227 5.00 0.00
5629 3.67 1.33
6032 2.45 2.55
6433 1.64 3.36
6836 1.10 3.90
7238 0.74 4.26
7641 0.49 4.51
8043 0.33 4.67
8444 0.22 4.78
8847 0.15 4.85
9249 0.10 4.90
9652 0.07 4.93
10053 0.05 4.95
10456 0.03 4.97
10858 0.02 4.98"""

# Processamento dos dados
linhas = dados.strip().split('\n')
tempo = []
tensao_resistor = []
tensao_capacitor = []

for linha in linhas:
    t, vr, vc = map(float, linha.split())
    tempo.append(t)
    tensao_resistor.append(vr)
    tensao_capacitor.append(vc)

# Configurao para melhor visualizao
plt.rcParams['figure.figsize'] = (12, 10)
plt.style.use('seaborn-v0_8')

# Grfico 1: Tenso no capacitor x tempo
plt.subplot(2, 2, 1)
plt.plot(tempo, tensao_capacitor, 'b-', linewidth=2, marker='o', markersize=4)
plt.title('Grfico 1: Tenso no Capacitor vs Tempo')
plt.xlabel('Tempo (ms)')
plt.ylabel('Tenso no Capacitor (V)')
plt.grid(True, alpha=0.3)

# Grfico 2: Tenso no resistor x tempo
plt.subplot(2, 2, 2)
plt.plot(tempo, tensao_resistor, 'r-', linewidth=2, marker='s', markersize=4)
plt.title('Grfico 2: Tenso no Resistor vs Tempo')
plt.xlabel('Tempo (ms)')
plt.ylabel('Tenso no Resistor (V)')
plt.grid(True, alpha=0.3)

# Grfico 3: Sobreposio dos dois grficos
plt.subplot(2, 1, 2)
plt.plot(tempo, tensao_capacitor, 'b-', linewidth=2, marker='o', markersize=4, label='Tenso no Capacitor')
plt.plot(tempo, tensao_resistor, 'r-', linewidth=2, marker='s', markersize=4, label='Tenso no Resistor')
plt.title('Grfico 3: Sobreposio - Tenses vs Tempo')
plt.xlabel('Tempo (ms)')
plt.ylabel('Tenso (V)')
plt.legend()
plt.grid(True, alpha=0.3)

# Ajustar layout para evitar sobreposio
plt.tight_layout()
plt.show()

# Estatsticas bsicas
print("Estatsticas dos dados:")
print(f"Tempo inicial: {tempo[0]} ms")
print(f"Tempo final: {tempo[-1]} ms")
print(f"Tenso mxima no capacitor: {max(tensao_capacitor):.2f} V")
print(f"Tenso mxima no resistor: {max(tensao_resistor):.2f} V")
print(f"Tenso final no capacitor: {tensao_capacitor[-1]:.2f} V")
print(f"Tenso final no resistor: {tensao_resistor[-1]:.2f} V")
```
